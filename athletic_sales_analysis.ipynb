```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
from pathlib import Path
ath_2020_path = Path("./Resources/athletic_sales_2020.csv")
ath_2021_path = Path("./Resources/athletic_sales_2020.csv")
ath_2020 = pd.read_csv(ath_2020_path)
ath_2021 = pd.read_csv(ath_2021_path)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
ath_full = pd.concat([ath_2020, ath_2021], axis = 0)
ath_full = ath_full.reset_index(drop=True)
ath_full
```

```python
# Check if any values are null.
cols = ath_full.columns
for col in cols:
    print(f"{col} {" " * (20 - len(col))} {sum(ath_full[col].isnull())}")
```

```python
# Check the data type of each column
ath_full.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
from datetime import datetime
ath_full["invoice_date"] = pd.to_datetime(ath_full["invoice_date"], yearfirst=True)
ath_full
```

```python
# Confirm that the "invoice_date" data type has been changed.
ath_full.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
relevant_columns = ["region", "state", "city", "units_sold"]
products_sold = ath_full[relevant_columns].groupby(by =["region", "state", "city"])[['units_sold']].sum()
products_sold = products_sold.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
sorted_total = products_sold.sort_values(["Total_Products_Sold"], ascending=False)
print(sorted_total.head(5))
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
products_sold = ath_full[relevant_columns].pivot_table(values="units_sold", index=["region", "state", "city"], aggfunc="sum")

# Rename the "units_sold" column to "Total_Products_Sold"
products_sold = products_sold.rename(columns={"units_sold":"Total_Products_Sold"})

# Show the top 5 results.
sorted_total = products_sold.sort_values(["Total_Products_Sold"], ascending=False)
print(sorted_total.head(5))
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.

relevant_columns = ["region", "state", "city", "total_sales"]
products_sold = ath_full[relevant_columns].groupby(by =["region", "state", "city"])[["total_sales"]].agg("sum")
# Rename the "total_sales" column to "Total Sales"
products_sold = products_sold.rename(columns={"total_sales":"Total Sales"})

# Show the top 5 results.
sorted_total = products_sold.sort_values(["Total Sales"], ascending=False)
print(sorted_total.head(5))
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
products_sold = ath_full[relevant_columns].pivot_table(values="total_sales", index=["region", "state", "city"], aggfunc="sum")
# Optional: Rename the "total_sales" column to "Total Sales"
products_sold = products_sold.rename(columns={"total_sales":"Total Sales"})

# Show the top 5 results.
sorted_total = products_sold.sort_values(["Total Sales"], ascending=False)
print(sorted_total.head(5))


```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_ath = ath_full.loc[ath_full['product'] == "Women's Apparel", :]
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
relevant_columns = ["retailer", "region", "state", "city", "units_sold"]
products_sold = womens_ath[relevant_columns].groupby(["retailer", "region", "state", "city"])[["units_sold"]].agg("sum")
products_sold = products_sold.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

# Show the top 5 results.
sorted_total = products_sold.sort_values(["Womens_Footwear_Units_Sold"], ascending=False)
print(sorted_total.head(5))

```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
relevant_columns = ["retailer", "region", "state", "city", "units_sold"]
products_sold = womens_ath[relevant_columns].pivot_table(values="units_sold", index=["region", "state", "city", "retailer"], aggfunc="sum")
products_sold = products_sold.rename(columns={"units_sold": "Womens_Footwear_Units_Sold"})

# Show the top 5 results.
sorted_total = products_sold.sort_values(["Womens_Footwear_Units_Sold"], ascending=False)
print(sorted_total.head(5))
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
day_most_wa = womens_ath.pivot_table(values="total_sales", index="invoice_date")

# Optional: Rename the "total_sales" column to "Total Sales"
day_most_wa = day_most_wa.rename(columns={"total_sales": "Total Sales"})

# Show the table.
day_most_wa
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
resamp = day_most_wa.resample("1D").sum()

# Sort the resampled pivot table in descending order on "Total Sales".
resamp = resamp.sort_values("Total Sales", ascending=False)
resamp
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
resamp = day_most_wa.resample("1W").sum()

# Sort the resampled pivot table in descending order on "Total Sales".
resamp = resamp.sort_values("Total Sales", ascending=False)
resamp
```
